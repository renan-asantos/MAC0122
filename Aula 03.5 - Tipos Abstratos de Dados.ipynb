#### O conceito de Tipo de Dados Abstrato (Abstract Data Type ou ADT)
É um conjunto de objetos e as operações sobre esses objetos. As operações definem uma interface entre o ADT e o resto do programa. A interface define o comportamento destas operações, o que elas fazem e não como fazem.
ADT especifica:

a) O tipo de dados armazenado (tipo conceitual, não int, bool, list, etc.)

b) As operações suportadas sobre esses dados

c) Os parâmetros dessas operações

d) O que cada operação faz e não como faz

Representação de um ADT em Python são as classes.

```python
class Fração:

    '''Define a classe Fração Ordinária.'''
    
    # Construtor da classe
    def __init__(self, topo = 0, base = 1):
        '''Cria uma instância de Fração.'''
        self.num = topo
        self.den = base
    # Retorna a soma de dois elementos da classe
    def Some_Fração(self, f1, f2):
        self.num = f1.num * f2.den + f1.den * f2.num
        self.den = f1.den * f2.den
    # Mostre uma fração na forma a/b
    def Mostre_Fração(self):
        print("{}/{}".format(self.num,self.den))

# Exemplos de uso da classe Fração
x = Fração(3, 4)
y = Fração(5, 12)
z = Fração()
# z = x + y
z.Some_Fração(x, y)
z.Mostre_Fração()
# z = 4/5 + 3/7
z.Some_Fração(Fração(4, 5), Fração(3, 7))
z.Mostre_Fração()
```

#### P3.5.5) Idem com as operações booleanas “está contido” com o operador < (__lt__), “contém” com o operador > (__gt__) e “igual” com o operador == (__eq__).
#### P3.5.6) Idem com a operação “pertinência” com o operador in (__contains__), “inclusão” de elemento com += (__iadd__) e exclusão com -=(__isub__).

```python
def Elimina_Repetidos(lista):    
    nova = []
    for x in lista:
        if not x in nova: nova.append(x)
    return nova

class Sset:

    ''' Define a classe Super Conjunto. '''
    # Construtor da classe
    def __init__(self, lista = []):
        ''' Cria uma instância de Sset. '''
        self.lset = Elimina_Repetidos(lista)
        self.lset.sort()
    # união (+)
    def __add__(s1, s2):
        # cria Sset com a concatenação de duas listas
        return Sset(s1.lset + s2.lset)
    
    # intersecção (*)
    def __mul__(s1, s2):
        inter = []
        l1 = len(s1.lset)
        l2 = len(s2.lset)
        if l1 > l2:
            for elem in s1.lset:
                if elem in s2.lset:
                    inter.append(elem)
        else:
            for elem in s2.lset:
                if elem in s1.lset:
                    inter.append(elem)
        return Sset(inter)
    
    # subtração (-)
    def __sub__(s1, s2):
        sub = []
        for elem in s1.lset:
            if elem not in s2.lset:
                sub.append(elem)
        return Sset(sub)
    
    # está contido (<)
    def __lt__(s1, s2):
        for elem in s1.lset:
            if elem not in s2.lset:
                return False
        return True
    
    # contém (>)
    def __gt__(s1, s2):
        for elem in s2.lset:
            if elem not in s1.lset:
                return False
        return True
    
    # igual (=)
    def __eq__(s1, s2):
        if len(s1.lset) == len(s2.lset):
            for elem in s1.lset:
                if elem not in s2.lset:
                    return False
            return True
        else:
            return False
    
    # pertinência (in)
    def __contains__(s1, x):
        if x in s1.lset:
            return True
        else:
            return False
   
    # inclusão de elemento (+=)
    def __iadd__(s1, x):
        s1.lset.append(x)
        return Sset(s1.lset)
    
    # exclusão de elemento (-=)
    def __isub__(s1, x):
        s1.lset.remove(x)
        return Sset(s1.lset)
    
    # tamanho do Sset
    def __len__(self):
        return len(self.lset)
    
    # Transforma elemento da classe em str para o print
    # Mostrar um '*' na frente para diferenciar do print(lista)
    def __str__(self):
        return '*' + str(self.lset)
    
x = Sset([1,2,3,4])
y = Sset([2,3])
print(x)
x+=5
print(x)
x-=5
print(x)
```

#### P3.5.7) Incremente a classe Vector vista anteriormente com as operações de subtração de vetores, produto escalar e produto vetorial.

```python
class Vector:
    
    '''Vetor multidimensional.'''
    
    def __init__ (self, d):  
        '''Cria vetor de dimensão d com zeros.'''
        self._coords = [0] * d
        
    def __len__ (self):
        '''Retorna a dimensão do vetor.'''
        return len(self._coords)
    
    def __getitem__ (self, j):
        '''Retorna a j-ésima coordenada.'''
        return self._coords[j]
    
    def __setitem__ (self, j, val):
        '''Define a j-ésima coordenada do vetor.'''
        self._coords[j] = val
        
    def __add__ (self, other):
        '''Retorna a soma de 2 vetores.'''
        if len(self) != len(other): # usa a função len desta classe
            raise ValueError("dimensões devem ser a mesma")
        result = Vector(len(self))
        for j in range(len(self)):
            result[j] = self[j] + other[j]
        return result
    
    def __sub__(self, other):
        '''Retorna a subtração de 2 vetores.'''
        if len(self) != len(other): # usa a função len desta classe
            raise ValueError("dimensões devem ser a mesma")
        result = Vector(len(self))
        for j in range(len(self)):
            result[j] = self[j] - other[j]
        return result
    
    def __mul__(self, other):
        '''Retorna o produto escalar entre 2 vetores.'''
        if len(self) != len(other): # usa a função len desta classe
            raise ValueError("dimensões devem ser a mesma")
        result = 0
        for j in range(len(self)):
            result += self[j]*other[j]
        return result
    
    def __pow__(v1, v2):
        '''Retorna o produto vetorial entre 2 vetores.'''
        if len(v1) != 3 or len(v2) != 3:
            raise ValueError("dimensões devem ser igual a 3")
        result = Vector(3)
        result[0] = v1[1]*v2[2] - v2[1]*v1[2]
        result[1] = v1[2]*v2[0] - v2[2]*v1[0]
        result[2] = v1[0]*v2[1] - v2[0]*v1[1]
        return result
    
    def __eq__ (self, other):
        '''Retorna True se vetores iguais.'''
        return self._coords == other._coords
    
    def __ne__ (self, other):
        '''Retorna True se vetores são diferentes.'''
        return not self == other # usa a definição de eq
    
    def imprime (self):
        '''Imprime o vetor.'''
        print(self._coords)
        
# testes
v1 = Vector(3)
v2 = Vector(3)
v1[0] = 1
v1[1] = 2
v1[2] = 3
v2[0] = 4
v2[1] = 5
v2[2] = 6
#v3 = v1 + v2 # add
#if v1 == v3 or v2 == v3: # eq
#    print("soma igual a um deles")
#else:
#    print("soma diferente")
#if v3 != Vector(3): # ne
#    print("Não é zero")
#v2.imprime()
v3 = v1**v2
v3.imprime()
```

```python
class Matriz:

    ''' Define a classe Matriz de n linhas por m colunas. '''
    
    def __init__(self, n, m):
        # inicialmente todos zerados
        #self.elem = [m * [0] for k in range(n)]
        self.elem = [m * [0]] * n
        
    # define a k-esima linha da Matriz
    def __setitem__(self, k, v):
        self.elem[k] = v[:]
    
    # tamanho das matrizes para consistência
    def dim(self):
        # linhas e colunas
        n, m = len(self.elem), len(self.elem[0])
        return (n, m)
    
    # soma duas matrizes de mesma dimensão
    def __add__(m1, m2):
        # consistência dos tamanhos
        if m1.dim() != m2.dim():
            raise TypeError("tamanhos diferentes")
        # linhas e colunas
        n, m = len(m1.elem), len(m1.elem[0])
        m3 = Matriz(n, m)
        for i in range(n):
            for j in range(m):
                m3.elem[i][j] = m1.elem[i][j] + m2.elem[i][j]
        return m3
    
    #Imprime a Matriz
    def __str__(self):
        lin = ''
        for i in range(len(self.elem)):
            for j in range(len(self.elem[i])):
                lin += str(self.elem[i][j]) + ' '
            lin += '\n'
        return lin

# testes
ma = Matriz(2, 3)
mb = Matriz(2, 3)
# define as linhas
ma[0] = [1, 2, 3]
ma[1] = [4, 5, 6]
mb[0] = [5, 5, 5]
mb[1] = [9, 9, 9]
# soma
mc = ma + mb
# imprima
print(ma)
print(mb)
print(mc)
```

Implemente as operações de subtração de matrizes (-) e multiplicação (*). Importante verificar a compatibilidade do tamanho das listas antes de realizar as operações.

```python
#Retorna o produto escalar entre duas listas
def produto_escalar(a,b):
    soma = 0
    for x,y in zip(a,b):
        soma += x*y
    return soma

class Matriz:

    ''' Define a classe Matriz de n linhas por m colunas. '''
    
    def __init__(self, n, m):
        # inicialmente todos zerados
        self.elem = [m * [0] for k in range(n)]
        # pode ser também [m * [0]] * n
        
    # define a k-esima linha da Matriz
    def __setitem__(self, k, v):
        self.elem[k] = v[:]
    
    # tamanho das matrizes para consistência
    def dim(self):
        # linhas e colunas
        n, m = len(self.elem), len(self.elem[0])
        return (n, m)
    
    # soma duas matrizes de mesma dimensão
    def __add__(m1, m2):
        # consistência dos tamanhos
        if m1.dim() != m2.dim():
            raise TypeError("tamanhos diferentes")
        # linhas e colunas
        n, m = len(m1.elem), len(m1.elem[0])
        m3 = Matriz(n, m)
        for i in range(n):
            for j in range(m):
                m3.elem[i][j] = m1.elem[i][j] + m2.elem[i][j]
        return m3
    
    #Subtrai duas matrizes
    def __sub__(m1, m2):
        # consistência dos tamanhos
        if m1.dim() != m2.dim():
            raise TypeError("tamanhos diferentes")
        # linhas e colunas
        n, m = len(m1.elem), len(m1.elem[0])
        m3 = Matriz(n, m)
        for i in range(n):
            for j in range(m):
                m3.elem[i][j] = m1.elem[i][j] - m2.elem[i][j]
        return m3
    
    #Multiplicar matrizes
    def __mul__(m1, m2):
        # consistência dos tamanhos
        lins1 = len(m1.elem)
        cols1 = len(m1.elem[0])
        lins2 = len(m2.elem)
        cols2 = len(m2.elem[0])
        if cols1 != lins2:
            raise TypeError("tamanhos diferentes para a multiplicação")
        m3 = Matriz(lins1, cols2)
        for i in range(lins1):
            for j in range(cols2):
                coluna = [m2.elem[K][j] for K in range(len(m2.elem))]
                m3.elem[i][j] = produto_escalar(m1.elem[i],coluna)
        return m3
    
    #Imprime a Matriz
    def __str__(self):
        lin = ''
        for i in range(len(self.elem)):
            for j in range(len(self.elem[i])):
                lin += str(self.elem[i][j]) + ' '
            lin += '\n'
        return lin

ma = Matriz(2, 1)
mb = Matriz(1, 2)
ma[0] = [1]
ma[1] = [2]
mb[0] = [3,4]
print(ma)
print(mb)
print(ma*mb)
```
